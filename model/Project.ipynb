```python
s="1221-2442"
t=s.split('-')
print(t)
print(len(t))
print("a list is created and list has two elemnts")
print(t[0])
print(t[1])
```

```python
#FUNCTION THAT TAKE INPUT STRING AND SEPARATE IT INTO 2 values SPlIT('-')
def convert_sqft_to_num(x):
    tokens=x.split('-')
    if len(tokens) ==2:
        return (float(tokens[0])+float(tokens[1]))/2
    try:
        return float(x) #convert it into floats
    except:
        return None
```

## OUTLIER DETECTION AND REMOVAL

```python
#suppose the threshold value is 1 bed for 300 sqft 
#if 6 bed for 600 sqft is given that means 1 bed will be for 100sqft (quite compact and unusual)
df5[df5.total_sqft/df5.bhk<300].head() #finds the outlier present 
```

```python
df7=df5[~(df5.total_sqft/df5.bhk<300)]    #just negate and remove the outlier present
df7.head()
```

#### 2nd outlier price/sqft

```python
# min=267 per sq ft is very low and unlikely and max is also same acting as outlier
for key,value in(df7.groupby('location')):
    print(key+'\n')
    print('##########')
    print(value)
    print('\n')
```

```python
#clearly we can see that location is grouped and for each location there are different price_per_sq_ft values
df7.price_per_sq_ft.describe()
```

```python
df7.shape
```

```python
# outliers are those which have price per sq ft values < mean - standard deviation and  
# price per sq ft values > mean + standard deviation 
def remove_price_per_sq_ft_outliers(df):
    df_out=pd.DataFrame()
    for key,values in df.groupby('location'):
        mean=np.mean(values['price_per_sq_ft'])
        SD=np.std(values['price_per_sq_ft'])
        reduced_df=values[(values['price_per_sq_ft']>(mean-SD)) & (values['price_per_sq_ft']<=(mean+SD))]
        df_out=pd.concat([df_out,reduced_df],ignore_index=True)
    return df_out
```

```python
df8=remove_price_per_sq_ft_outliers(df7)
df8.shape
```

## FINDING BEST MODEL USING GridSearchCV

```python
from sklearn.model_selection import GridSearchCV

from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

def find_best_model_using_gridsearchcv(X,y):
    algos = {
        'linear_regression' : {
            'model': LinearRegression(),
            'params': {
                'normalize': [True, False]
            }
        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1,2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion' : ['mse','friedman_mse'],
                'splitter': ['best','random']
            }
        }
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        gs =  GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(X,y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(scores,columns=['model','best_score','best_params'])

find_best_model_using_gridsearchcv(x,y)
```
